```python
#remove the "$\t, and white space"
#X["SALE PRICE"][4][3:].rstrip()

A = []
for i in range(len(X["SALE PRICE"])):
    price_no_commas = ""
    for ch in X["SALE PRICE"][i]:
        if ch != ",":
            price_no_commas += ch
    A.append(price_no_commas[3:].rstrip())
   

```

```python
df["SALE PRICE NEW"] = A

df.drop("SALE PRICE", axis=1, inplace = True)
#note that setting axis = 1 makes .drop() remove columns 
#          setting axis = 0 makes .drop() remove rows

#drop can accept a list as its first argument, dropping either a list of column names (strings)
#or a list of row indexes (ints)

df.head()
```

```python
#write a code that drops a row iff the sale price new entry is "-"

#for loop appends a list of index values where NA prices occur
index = []
i = 0
for price in df["SALE PRICE NEW"]:
    if not price.isdigit():
        index.append(i)
    i += 1


#the .drop() method takes a list of values to remove, and an axis = 0 (since we're removing rows)
df.drop(index, axis=0, inplace=True)
```
