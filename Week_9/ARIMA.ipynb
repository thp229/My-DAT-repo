## Auto-Regressive and Moving Average Processes

```python
# stationarity -> constant serial mean/constant serial variance 

#DIFFERENCING transforms a non-random, highly correlated time series,
#with nonconstant serial mean, into a stationary sequence
##(however differencing does not always create nonconstant variance) 
```

```python
#for stock time series, differencing is the same thing as finding the log-rate of return 
```

```python
#AR - I - MA 

#I = 'the number of differencing terms' 

##what about the AR and the MA?
```

```python
# ARIMA(p,d,q) model

# p = number of autoregressive terms (the number of units of time in your lookback)
# d = number of times you difference your dataset
# q = number of moving average terms 
```

```python
#AR is linear regression where your regressing against the previous values of the actual series
##its a regression of the sequence upon itself
```

```python
#ex of an AR(2) process
#you look at the last two values in the sequence that you have
#weight them accordingly and add a noise constant

#i.e. 
#prediction_t = noise_constant + weight1*Realizedvalue_(t-1) + weight2*Realizedvalue_(t-2) 
```

```python
#previously occuring values all by themselves are not enough to make good predictions however
#big question: 
## how do we choose 'p' and how do we choose [weight1, weight2, .. ,weightP]?
```

```python
#also this model is missing the natural statistical noise (i.e. the random walk element)
#this is where the MA component comes in to play 

#the MA(q) refers to the previously observed values of statistical noise in the dataset
##works alot like AR, but now we are regressing upon previously observed noise terms 

#with ARIMA, you get a blend of previously observed values, and previously observed random noise
#the MA(q) part is really where the stationarity requirement comes into play
```

```python
#MA(q)
#q is the number of previous noise terms we regress against 
#noise is represented by the residuals from your previous estimate
```

```python
#Q: how do we find our values of P and Q
#A: this is derived from our autocorrelation plots 

#really p and q denote the number of statistically significant lags you have in your dataset 
```

```python
path = "/Users/theodoreplotkin/desktop/postmalone/GA_Data_Science/DAT-06-24/class material/Unit 4/Class 18/data/AirPassengers.csv"
```

```python
import pandas as pd
```

```python
data = pd.read_csv(path)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

stationary_data = data["Passengers"].diff().diff(periods=12)[13:]
x = plot_acf(stationary_data, lags = 40)
```

```python
#note autocorrelation is strongest at the very earliest lags (lag = 1~2)
#note: disregard the 0th value, lag 0 will always be 1 
#use the number of statistically significant lags as your value of q
```

```python
y = plot_pacf(stationary_data,lags = 40)
```

```python
#the partial autocorrelation informs us of how we choose the lag p
#this shows us there is clearly a seasonal component in our series 
##suggests we should include a seasonal component in our AR terms 
```

```python
#so lets fit an ARIMA(1,1,1) model (not accounting for seasonality right now)

from statsmodels.tsa.arima_model import ARIMA
```

```python
data.tail()
```

```python
model = ARIMA(data["Passengers"],order = (1,1,1))
results = model.fit()
print(results.summary())
```

```python
results.forecast() #note the .forecast() method

#first number is the forecast
#second number is the std of the forecast
#third number is the 95% confidence interval
```
