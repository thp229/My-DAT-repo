```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

drinks = pd.read_csv("/Users/theodoreplotkin/desktop/postmalone/GA_Data_Science/DAT-06-24/class material/Unit 2/data/drinks.csv")
```

```python
#renaming, adding and removing columns from a dataframe
print(drinks.columns)
```

```python
#.rename function allows us to rename columns

drinks.rename(columns={"beer_servings":"beer", "wine_servings":"wine"})

#rename takes the following input:
#columns = {a dictionary where they key is the original column name and the value is the new column name}

#this returns a view of the dataframe, but doesn't actually change the dataframe
```

```python
#in order to make the change permanent, we must add the inplace argument and set it to True
drinks.rename(columns={"beer_servings":"beer", "wine_servings":"wine"}, inplace = True)
```

```python
drinks.head() #note that the changes are now inplace 
```

```python
# another way to change column names that applies when you load in the data initially::

#create a list of names of matching length

#drink_cols = ["country", "beer", "spirit", "wine", "liters","continent"]

#drinks = pd.read_csv(filepath, header = 0, names =drink_cols)
            #header = 0 means that the first row of the dataframe will be the header
        
            #names = drink_cols set the column names equal to the list
```

```python
#can also change column names after the file has already been read into python as follows

drink_cols = ["country", "beer", "spirit", "wine", "liters","continent"]

drinks.columns = drink_cols 
```

```python
drinks.head()
```

```python
#removing columns from a dataframe 

#we use the df.drop() method
#by default, .drop() drops rows <it defaults on axis 0>

drinks.drop("liters", axis = 1) #we must specify axis = 1 so that .drop applies across a column

#setting inplace = True actually changes the underlying dataframe
#if we do not include inplace = True, we simply return a view of the df without altering the underlying df
                        
#can also pass in a list of column names and .drop() will remove all these columns at once
```

```python
#drinks = drinks.drop("liters", axis = 1)
#this is equivalent to setting inplace = True, since it actually overwrites the value of the dataframe with the subset
```

```python
drinks.continent.value_counts()
#the .value_counts() method excludes NA
```

```python
#we can pass in the following argument into .value_counts() to pickup the missing values

drinks.continent.value_counts(dropna = False)
#now we are listing the number of NaN's in the continent column 
```

```python
#another important method for missing values is .isnull()

#.isnull() loops thru the array and returns True if that value is NaN and false otherwise
drinks.continent.isnull()
```

```python
drinks.continent.isnull().sum() #essentially sums up the indicator for missing values
                                #True == 1 and False == 0 
```

```python
drinks.isnull().sum() #tells us the number of missing values in each column
```

```python
#the complement method to .isnull() is .notnull()

drinks.continent.notnull() #loops thru the array and assigns True if the value is not NaN
```

```python
drinks.continent.notnull().sum()
```

```python
#if we want to actually remove NaN rows

drinks[drinks.continent.notnull()].head() #presents a view of the df with the NaN's removed

#note that the axis is by default 0, which corresponds to the rows
#so its as if we are passing drinks a list of index values corresponding to rows which are not NaN's
```

```python
print(drinks.isnull().sum())
```

```python
drinks.isnull().sum().plot(kind = "bar") #note we are using the .plot() method 
                                        
plt.title("Number of null values per column")
```

```python
#dropping missing values with the .dropna() method

drinks.dropna() #drop a row if ANY values are missing from any column in that row -- can be dangerous!

#we can of course pass in arguments to modify .dropna()'s functionality 
```

```python
#the .fillna() method allows us to replace a NaN with a value of your choice

drinks.continent.fillna(value = "NA")
#again this simply returns a view of the dataframe, to actually modify the dataframe 
#with our changes we must add the argument inplace and set inplace = True 
```

```python
#drinks.continent.fillna(value = "NA", inplace = True) actually overwrites the dataframe with the changes
```

```python
#Split-Apply-Combine method

#split: we group data by continent
#apply: we apply the .mean() method 
#combine: 

drinks.head()
```

```python
#key method is .groupby()
#takes every single unique value in the continent column, and find the average value within a column of choice

#for each continent, calculate the mean beer servings
drinks.groupby("continent").beer.mean()

#group the entire dataframe by continent 
    #then within the beer column, find the mean for each group 
```

```python
drinks.groupby("continent").mean()

#now we are finding the mean of each column <beer, spirit, wine, liters> within each continent category
```

```python
#we can also apply other methods to .groupby() such as .describe()
drinks.groupby("continent").beer.describe()
```

```python
#.agg takes a list of summary stat features and returns only those specific summary stats

drinks.groupby("continent").beer.agg(["count", "mean", "min", "max"])
```

```python
#.sort_values() allows us to sort by a specific column in the summary stats 
drinks.groupby("continent").beer.agg(["count", "mean", "min", "max"]).sort_values("mean")
```

```python
drinks.groupby("continent").describe()
```
