```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("/Users/theodoreplotkin/desktop/postmalone/GA_Data_Science/DAT-06-24/class material/Unit 2/data/titanic.csv")
```

```python
df.head()
```

```python
#Kernel Density Estimate (KDE)
#a smoothed histogram function (as opposed to .hist() method which discretely bins the data)

#note the .plot() method which takes a kind argument
df["Age"].plot(kind="kde") #displays the probability density function of the data
```

```python
#Scatter plots using the .plot() method

df.plot("Age", "Fare", kind="scatter") #must specify the X and Y variables as well as kind = "scatter"
                                       #we can customize the range with xlim, ylim arguments
```

```python
#scatterplots tend to look nicer in Seaborn
```

```python
#Hexbins with the .plot() method

df.plot("Age", "Fare", kind="hexbin")
```

```python
#subplots are often easier in Seaborn
#but this is how you do them in matplotlib.pyplot

fig, ax = plt.subplots(1,2) #1 row, 2 columns of plots

#note that the subplot function returns two values and thus must be called with two variables! 

#ax is the nested list of axis objects 
#fig is the actual figure frame seen below
```

```python
ax #these are the axis objects for the above plots, stored in lists
        #if we had multiple rows, these would be nested lists
```

```python
fig, ax = plt.subplots(1,2)
df[df.Sex == "female"].plot("Age","Fare",kind="scatter",ax = ax[0], color = "red", label = "Females")
df[df.Sex == "male"].plot("Age","Fare",kind="scatter", ax=ax[1], color = "blue", label = "Males")
#note that we call the .plot() function on the relevent subset of the dataframe 

#label gives us our legend
#note that we must specify which axis object we want to assign the plot to, this is done with the ax = ax[index] argument
```

```python
fig, ax = plt.subplots(1,2, sharex = True, sharey = True, figsize = (12,4))
df[df.Sex == "female"].plot("Age","Fare",kind="scatter",ax = ax[0], color = "red", label = "Females")
df[df.Sex == "male"].plot("Age","Fare",kind="scatter", ax=ax[1], color = "blue", label = "Males")

#sharex makes all plots share the same x axis scale
#sharey makes all the plots share the same y axis scale
#figsize sets the figures aspect ratios
```
