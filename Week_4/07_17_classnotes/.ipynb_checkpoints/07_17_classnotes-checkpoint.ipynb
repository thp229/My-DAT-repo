```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("/Users/theodoreplotkin/desktop/postmalone/GA_Data_Science/DAT-06-24/class material/Unit 2/data/titanic.csv")
```

```python
df.head()
```

```python
df.groupby("Sex")[["Age","Fare"]].mean().plot(kind="bar")
plt.title("Average Age and Fare by Passenger Gender")
plt.show()
```

```python
df["Embarked"].value_counts().plot(kind="pie")
plt.show()
```

```python
#Kernel Density Estimate (KDE)
#a smoothed histogram function (as opposed to .hist() method which discretely bins the data)

#note the .plot() method which takes a kind argument
df["Age"].plot(kind="kde") #displays the probability density function of the data
plt.show()
```

```python
#Scatter plots using the .plot() method

df.plot("Age", "Fare", kind="scatter") #must specify the X and Y variables as well as kind = "scatter"
                                       #we can customize the range with xlim, ylim arguments
```

```python
#scatterplots tend to look nicer in Seaborn
```

```python
#Hexbins with the .plot() method

df.plot("Age", "Fare", kind="hexbin")
```

```python
#subplots are often easier in Seaborn
#but this is how you do them in matplotlib.pyplot

fig, ax = plt.subplots(1,2) #1 row, 2 columns of plots

#note that the subplot function returns two values and thus must be called with two variables! 

#ax is the nested list of axis objects 
#fig is the actual figure frame seen below
```

```python
ax #these are the axis objects for the above plots, stored in lists
        #if we had multiple rows, these would be nested lists
```

```python
fig, ax = plt.subplots(1,2)
df[df.Sex == "female"].plot("Age","Fare",kind="scatter",ax = ax[0], color = "red", label = "Females")
df[df.Sex == "male"].plot("Age","Fare",kind="scatter", ax=ax[1], color = "blue", label = "Males")
#note that we call the .plot() function on the relevent subset of the dataframe 

#label gives us our legend
#note that we must specify which axis object we want to assign the plot to 
#     this is done with the ax = ax[row_index][col_index] argument
```

```python
fig, ax = plt.subplots(1,2, sharex = True, sharey = True, figsize = (12,4))
df[df.Sex == "female"].plot("Age","Fare",kind="scatter",ax = ax[0], color = "red", label = "Females")
df[df.Sex == "male"].plot("Age","Fare",kind="scatter", ax=ax[1], color = "blue", label = "Males")

#sharex = True, makes all plots share the same x axis scale
#sharey = True, makes all the plots share the same y axis scale
#figsize = (a,b) sets the figures aspect ratios
```

```python
fig, ax = plt.subplots(2,2)
ax #note the ax variable returns a nested list, so to assign plots we must use pair indexing
```

```python
#enter the Seaborn graphing library, built ontop of matplotlib.pylot 
#userfriendly way to make good-looking charts
##good for heatmaps, regression plots, pairplots, etc instantly

#primary chart types are
# catplot, relplot, implot/regplot, displot/jointplot, pairplot
```

```python
#CATPLOT is useful for categorical data, with 9 variants for visual diversity 

#default category plot, a strip-plot
sns.catplot(x="Pclass", y="Age", data=df) #KEY POINT: the X argument is a categorical data-type
3                                          #           the Y argument is a cts data-type
```

```python
#specifying the kind argument gives us the subvariants 

sns.catplot(x="Pclass", y="Age", kind="swarm", data=df)
#setting kind = "swarm" shows us more precisely the density of the distribution of various values 
```

```python
#there are other arguments 
#--strip
#--box
#--boxen
#--violin
#--point
#--bar
#--count

sns.catplot(x="Pclass", y="Age", data=df, kind = "box") #setting kind = "box" gives us a traditional box-plot
```

```python
sns.catplot(x="Pclass", y="Age", data=df, kind = "boxen") #setting kind = "boxen" gives us a telescoping box-plot
```

```python
sns.catplot("Pclass", data=df, kind = "count") #setting kind = "count" gives us a barchart representing the frequency
#                                                of occurance of each category

                                               #note if kind = "count", then we only need to pass in 1 data column 
                                                    #works best on catagorical data columns 


#basically yields the same result as 
#   df["Pclass"].value_counts().plot(kind="bar")
```

```python
#other useful arguments in Seaborn:

#hue and col/row
```

```python
sns.catplot(x="Pclass", y="Age", kind="swarm", data=df, hue="Sex")

#note the use of the hue argument
    #hue can be set to a categorical data type and will automatically break the plot up by color based on which points
    #fall into which category 
```

```python
sns.catplot(x="Pclass", y="Age", kind="swarm", data=df, col="Sex")

#note the use of the col argument
    #col can be set to a categorical data type and will automatically create a column
    #of two seperate plots for each subset of the data
```

```python
sns.catplot(x="Pclass", y="Age", kind="swarm", data=df, row="Sex")

#note the use of the row argument
    #does the same as the col argument, but breaks the plots apart row-wise
    
#this offers an easier way to make subplots than using matplotlib and the axis-object
```

```python
sns.catplot(x="Pclass", y="Survived", kind="bar", data=df,hue="Sex")
plt.title("Survival Rate by PAssnger Class/Sex")

#to add titles with seaborn simply use .title() method from matplotlib.pyplot
```

```python
#quantiative charts
##lmplots <-- linear-model plots, automatically adds line of best fit with confidence bound to scatterplots

##distribution/jointplots 

##pairplots/heatmaps
```

```python
#to draw the line of best fit thru two categorical variables use .lmplot() method

sns.lmplot(x="Age", y="Fare", data=df) #KEYPOINT: both X and Y data-types must be cts
```

```python
#hue, and col/row also applies to .lmplot()

sns.lmplot(x="Age", y="Fare", hue="Sex",data=df)
```

```python
sns.lmplot(x="Age", y="Fare", col="Sex",data=df)
```

```python
#a better looking histogram with .distplot() with a Kernel Density Estimate overlayed

df["Age"]=df["Age"].fillna(29) #we're simply filling in missing values, note the .fillna() method

sns.distplot(df.Age) 
```

```python
#.jointplot() gives us a scatterplot with histogram overlays for each variable

sns.jointplot(x="Age", y="Fare", data=df[df.Sex=="male"]) #note we're calling the method on only the male subset
```

```python
#NOTE: Hue and Col/Row splits do not work for .distplots()
```

```python
#Specialty plots with pairplots and heatmaps
sns.pairplot(df)

#basically a scatter-matrix of all variables, with diagonal elements as histograms
```

```python
sns.pairplot(df,x_vars=["Pclass","Age","Fare","SibSp"],y_vars=["Age"])

#note the use of x_vars and y_vars allows us to specify specific columns to apply pairplot to
```

```python
df.corr() #automatically returns the correlation matrix with the .corr() method applied to the dataframe
```

```python
#we can visualize the correlation matrix with the .heatmap() method
sns.heatmap(df.corr())
```
